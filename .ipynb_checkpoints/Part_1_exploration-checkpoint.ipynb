### Introduction
The ford Gobike dataset includes information about individual rides made in a bike sharing system covering the greater San Francisco Bay area. The dataset has 16 columns and 183412 rows. The columns are:

 - start_time - starting time
 - end_time - ending time
 - start_station_id - Starting station id
 - start_station_name - The name of the starting station
 - start_station_latitude - Starting station in latitude
 - start_station_longitude - Starting station location in longitude
 - end_station_id - Ending station id
 - end_station_name - Ending station name
 - end_station_latitude -Ending station location in latitude
 - end_station_longitude - Ending station location in longitude
 - bike_id - bike id
 - user_type - User type either Subscriber or Customer
 - member_birth_year - Member birth year 
 - member_gender - Member gender
 - bike_share_for_all_trip - is the bike used for round trips?

## Cleaning
  - In this section, I will address the quality issues

```python
# Change the start and end_time columns from object to dateformat
def date_conversion(start,end):
    global copy_b
    copy_b[start]=pd.to_datetime(copy_b[start])
    copy_b[end]=pd.to_datetime(copy_b[end])
    
    #create new columns for starting and ending time
    copy_b['Starting_time']=copy_b[start].dt.strftime('Y%-m%-d%')
    copy_b['Ending_time']=copy_b[end].dt.strftime('Y%-m%-d%')
    
    #separate the hours
    copy_b['Start_hour']=copy_b[start].dt.hour
    copy_b['End_hour']=copy_b[end].dt.hour
    
    #separate the days
    copy_b['Start_day']=copy_b[start].dt.strftime('%a')
    copy_b['End_day']=copy_b[end].dt.strftime('%a')
    
    #separate the months
    copy_b['Start_month']=copy_b[start].dt.strftime('%B')
    copy_b['End_month']=copy_b[end].dt.strftime('%B')
    
    copy_b=copy_b.drop(columns=[start,end])
  

    
```

```python
#call the function date_conversion
date_conversion('start_time','end_time')
```

```python
#check for the changes
copy_b.head()
```

```python
# drop rows with missing values
copy_b.dropna(axis=0)
```

### Q3. Which day has the highest number of riders?

```python
#convert the column Start_day to a categorital type
classes=['Sun','Mon','Tue','Wed','Thu','Fri','Sat']

vclasses=pd.api.types.CategoricalDtype(ordered=True, categories=classes)

copy_b['Start_day']=copy_b['Start_day'].astype(vclasses);
```

```python

sns.countplot(data=copy_b,x='Start_day',color=color)
plt.title('Number of rides in a week')
plt.xlabel('Days of the week')
plt.ylabel('Number of rides')
plt.xticks(np.arange(0,6+1,1),labels=classes);
```

### From the bar chart, Thursday has the highest number of riders and most people ride from Monday to Friday. The weekend has less riders compared to the weekdays(Mon-Fri).

```python
#let's visualize the above bar chart in a different way
group_data=copy_b.groupby('Start_day')['user_type'].value_counts()
group_data
```

```python
#Use series.reset_index() to a series into a dataframe object
group_data=group_data.reset_index(name='count')
group_data
```

```python
#divide the count column with 100 to make it smaller
group_data['count']=group_data['count']/100
```

```python
group_data
```

```python
# Use DataFrame.pivot() to rearrange the data, to have vehicle class on rows
group_data=group_data.pivot(index='Start_day',columns='user_type',values='count')
group_data
```

```python
#plot the heatmap
sns.heatmap(group_data,annot=True,fmt='.2f',cmap='vlag');
```

### Still the same results.

### Q5. What is the average duration for riders during the week?

```python
average_duration = copy_b.groupby('Start_day').duration_sec.mean()

# plot a line plot
plt.figure(figsize=[10,5])
sns.lineplot(x=average_duration.index, y=average_duration.values, marker='o')
plt.xticks(np.arange(0,7))
plt.ylabel('Average duration(seconds)')
plt.title('Average duration for riders on days of the week');
```

### The average duration is high on the weekends compared to (Mon-Fri).

# Multivariate Exploration


### Q1. What is the average duration for the week for each user

```python
combined_duration = copy_b.groupby(['Start_day','user_type']).duration_sec.mean()
combined_duration=combined_duration.reset_index()
combined_duration
```

```python
# plot a line plot
plt.figure(figsize=[10,5])
sns.lineplot(x=combined_duration['Start_day'], y=combined_duration['duration_sec'],hue=combined_duration['user_type'], marker='o')
plt.xticks(np.arange(0,7))
plt.ylabel('Average duration(seconds)')
plt.title('Average duration for User types');
```

### Well,it is clear that customers lasted longer than subscribers on average

### Q2. What is the relationship between gender,user type and day of the week

```python
g = copy_b.groupby(['Start_day', 'member_gender', 'user_type']).size().reset_index(name='count')
g = g.pivot(index='Start_day', columns=['member_gender', 'user_type'], values = 'count').dropna()
g
```

```python
sns.heatmap(g, annot=True, fmt='.0f', cmap='vlag')
plt.xlabel("Gender - User Type", fontsize=12)
plt.ylabel("Day of the Week", fontsize=12)
plt.title('Distribution for gender,user type and day of the week');
```

### The number of riders tend to be high between (Mon-Fri) and most of the riders are male subscribers.

 - On average customers lasted than customers
 - The number of riders tend to be high between (Mon-Fri) and most of the riders are male subscribers
